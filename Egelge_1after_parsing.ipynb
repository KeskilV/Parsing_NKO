### Using pattern 

```python
pattern_miuz0 = r"""(?ix)(?P<n>\d{1,4})\s+(?P<gem>\w+)\s+огранки\s+"
                    (?P<form>[\W\w]+)"[\s\w]*(?:граней)?\s?-?\s?(?P<graney>\d*)?"""
```

```python
pattern_egelge = r"""(?ix)(?P<n>\d{1,4})\s+(?P<gem>\w+)\s+
                    (?P<form>\w+\s+\d+)\s+ 
                    (?P<carat>\d+.\d+)Ct\s+  
                    (?P<col>\d+-?\d?)/(?P<cla>\d+)\s? 
                    (?P<po>[\W\w]{1})\s* """
```

```python
t = '1 Бр Кр 57 0.009Ct 3/6А'
#'8 Бр Кр 57 0.050Ct 7/9А черный'
#'51 Бр Кр 17 0.200Ct 2/3А или 56 Бр Кр 17 0.220Ct 2/3А'
```

```python
re.findall(pattern_egelge,t)
```

```python
re.search(pattern_egelge,t).group('n')
```

```python
re.search(pattern_egelge,t).group('gem')
```

```python
#DRY
dic_d = {'ШТУК':'n','gem':'gem','ОГРАНКА':'form','КАРАТ':'carat','ПО':'po','Ц':'col','Д':'cla'}
```

```python
for k in dic_d.keys():
    x = vstavki.vstavki_list1.apply(lambda x: re.search(
    pattern_egelge,x).group(dic_d[k]) if re.search(
    pattern_egelge,x) else 0)
    print(k, dic_d[k],'\n', x.unique())
    vstavki[k] = x
```
